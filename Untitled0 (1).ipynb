```python
import os
import random
import shutil
from glob import glob
from pathlib import Path
from PIL import Image

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report, confusion_matrix

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import VGG16, ResNet50, MobileNetV2
from tensorflow.keras.layers import GlobalAveragePooling2D, Dense, Dropout
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import ModelCheckpoint, ReduceLROnPlateau, EarlyStopping
```

```python
# Path hasil ekstraksi
base_path = Path("ham1000")

# Cari metadata CSV
metadata_file = list(base_path.rglob("GroundTruth.csv"))[0]
print("Metadata file ditemukan:", metadata_file)

# Load metadata
df = pd.read_csv(metadata_file)
print("Jumlah data:", len(df))
df.head()

# Determine the label for each row by finding the column with the maximum value
df['dx'] = df[['MEL', 'NV', 'BCC', 'AKIEC', 'BKL', 'DF', 'VASC']].idxmax(axis=1)

# Cek kolom dan label unik
print("\nKolom yang tersedia:", df.columns.tolist())
print("\nDistribusi label:")
print(df['dx'].value_counts())

# Plot distribusi label
import seaborn as sns # Add seaborn import
plt.figure(figsize=(8,4))
sns.countplot(x='dx', data=df, order=df['dx'].value_counts().index)
plt.title("Distribusi Kelas HAM10000")
plt.xticks(rotation=45)
plt.show()
```

```python
# Cek contoh gambar
img_dir = list(base_path.rglob("images"))[0]  # folder dengan gambar
print("Folder gambar:", img_dir)



sample_images = df.sample(6, random_state=42)
plt.figure(figsize=(12,6))
for i, (_, row) in enumerate(sample_images.iterrows()):
    img_path = img_dir / f"{row['image']}.jpg"
    img = Image.open(img_path)
    plt.subplot(2,3,i+1)
    plt.imshow(img)
    plt.title(row['dx'])
    plt.axis('off')
plt.show()
```

```python
# Buat folder dataset terstruktur
base_dataset = Path("ham1000_split")
train_dir = base_dataset / "train"
val_dir = base_dataset / "val"
test_dir = base_dataset / "test"

for d in [train_dir, val_dir, test_dir]:
    d.mkdir(parents=True, exist_ok=True)

# Tambahkan kolom path gambar
df['path'] = df['image'].apply(lambda x: str(img_dir / f"{x}.jpg"))
df = df[df['path'].apply(os.path.exists)]

# Split train-val-test
train_df, temp_df = train_test_split(df, test_size=0.2, stratify=df['dx'], random_state=42)
val_df, test_df = train_test_split(temp_df, test_size=0.5, stratify=temp_df['dx'], random_state=42)

print("Train:", len(train_df), "Val:", len(val_df), "Test:", len(test_df))

# Fungsi untuk menyalin gambar ke folder sesuai kelas
def copy_images(data, split_dir):
    for _, row in data.iterrows():
        class_dir = split_dir / row['dx']
        class_dir.mkdir(parents=True, exist_ok=True)
        dst = class_dir / f"{row['image']}.jpg"
        if not dst.exists():
            shutil.copy(row['path'], dst)

copy_images(train_df, train_dir)
copy_images(val_df, val_dir)
copy_images(test_df, test_dir)
print("Dataset tersusun dalam folder train/val/test ")
```

```python
IMAGE_SIZE = (224, 224)
BATCH_SIZE = 32

# Dataset pipeline
train_ds = tf.keras.utils.image_dataset_from_directory(
    train_dir, label_mode="categorical", image_size=IMAGE_SIZE, batch_size=BATCH_SIZE)
val_ds = tf.keras.utils.image_dataset_from_directory(
    val_dir, label_mode="categorical", image_size=IMAGE_SIZE, batch_size=BATCH_SIZE)
test_ds = tf.keras.utils.image_dataset_from_directory(
    test_dir, label_mode="categorical", image_size=IMAGE_SIZE, batch_size=BATCH_SIZE, shuffle=False)

class_names = train_ds.class_names
num_classes = len(class_names)
print("Kelas:", class_names)

# Prefetch for speed
AUTOTUNE = tf.data.AUTOTUNE
train_ds = train_ds.prefetch(AUTOTUNE)
val_ds = val_ds.prefetch(AUTOTUNE)
test_ds = test_ds.prefetch(AUTOTUNE)

# Augmentasi
data_augmentation = keras.Sequential([
    layers.RandomFlip("horizontal_and_vertical"),
    layers.RandomRotation(0.2),
    layers.RandomZoom(0.2),
    layers.RandomContrast(0.2)
])
```

```python
# Fungsi builder model umum
def build_model(base_model_fn, preprocess_input, name):
    base = base_model_fn(
        include_top=False,
        weights="imagenet",
        input_shape=IMAGE_SIZE + (3,),
        pooling='avg')
    base.trainable = False

    inputs = keras.Input(shape=IMAGE_SIZE + (3,))
    x = data_augmentation(inputs)
    x = preprocess_input(x)
    x = base(x, training=False)
    x = layers.Dense(256, activation='relu')(x)
    x = layers.Dropout(0.5)(x)
    outputs = layers.Dense(num_classes, activation='softmax')(x)
    model = keras.Model(inputs, outputs, name=name)
    return model
```

```python
# Model candidates
models_info = {
    "VGG16": (tf.keras.applications.VGG16, tf.keras.applications.vgg16.preprocess_input),
    "ResNet50": (tf.keras.applications.ResNet50, tf.keras.applications.resnet.preprocess_input),
    "MobileNetV2": (tf.keras.applications.MobileNetV2, tf.keras.applications.mobilenet_v2.preprocess_input)
}

histories = {}
for name, (base_fn, preproc) in models_info.items():
    print(f"\n=== Training {name} ===")
    model = build_model(base_fn, preproc, name)
    model.compile(optimizer=keras.optimizers.Adam(1e-4),
                  loss="categorical_crossentropy",
                  metrics=["accuracy"])
    history = model.fit(train_ds, validation_data=val_ds, epochs=20)
    histories[name] = history
    model.save(f"{name}_ham10000.h5")

# Plot hasil
plt.figure(figsize=(15,5))
for i, name in enumerate(histories.keys(), 1):
    plt.subplot(1,3,i)
    plt.plot(histories[name].history['val_accuracy'], label='val_acc')
    plt.plot(histories[name].history['accuracy'], label='train_acc')
    plt.title(name)
    plt.legend()
plt.show()
```
